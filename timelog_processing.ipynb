#### Processing the timelog.txt file to the following form:

```python
"""
Break                      65 minutes    6%
Dictionaries               15 minutes    1%
Exercises                 340 minutes   35%
Functions                  30 minutes    3%
Introduction              100 minutes   10%
Lists                      60 minutes    6%
Lists and Tuples           60 minutes    6%
Lunch Break               150 minutes   15%
Numbers and strings        55 minutes    5%
Solutions                  95 minutes    9%
"""
```

```python
with open('timelog.txt','r') as f:
    timetable = f.read().splitlines()
    
# Splittting data into two lists: hours and activities
hours = [x[0:5] for x in timetable if x != '']
activity = [x[5:] for x in timetable if x[5:] != ' End']

# Removing '' from activity
activity = [x[1:] for x in activity if x != '']

# Creating variable time that will hold the hour intervals of acitivities
time = []

for n,i in enumerate(hours):
    # Exception of last element
    if n == len(hours)-1:
        pass
    # Exception for next day
    elif n == (len(hours)/2)-1:
        pass
    else:
        time.append('{}-{}'.format(i,hours[n+1]))
        
# Zipping together time intervals and acitivties       
modded_timetable = [c + ' ' + a for c, a in zip(time,activity)]
       
# Creating list that will hold time variables in minutes
minutes_time = []

# Extracting time by slicing strings to hours and minutes,
# Hours * 60 + minutes = total minutes
for t in time:
    min_t = t.split('-')[0]
    max_t = t.split('-')[1]
    minutes_time.append((int(max_t[0:2])*60 + int(max_t[3:]))-
                        (int(min_t[0:2])*60 + int(min_t[3:])))

# Creating variable to hold total time (for percentage calcsulations)
sum_time = sum(minutes_time)

# Creating list of tuples, where tuple is: (minutes, activity)
tuple_ma = list(zip(minutes_time,activity))

# Creating empty dictionary and by iterating through tuple_ma 
# Adding activities as keys and then incrementing values of those keys
total_time_dict = {}
for m,a in tuple_ma:
    if not a in total_time_dict:
        total_time_dict[a] = m
    else:
        total_time_dict[a] += m
        
# PRINING
# Printing timetable
for n,i in enumerate(modded_timetable):
    # inserting blank line in the middle for empty line division
    if n == len(modded_timetable)/2:
        print('')
    else:
        print(i) 
        
# Printing empty line between two blocks
print('')

# Printing sums of activities
for key,value in sorted(total_time_dict.items()):
    print('{:<20} {:>8} minutes {:>6} %'.format(key,value,round(value/sum_time * 100,2)))
```
